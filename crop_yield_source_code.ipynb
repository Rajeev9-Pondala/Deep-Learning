## Training Data:

The dataset will be split to two datasets, the training dataset and test dataset. The data is usually tend to be split inequality because training the model usually requires as much data-points as possible.The common splits are 70/30 or 80/20 for train/test.

The training dataset is the intial dataset used to train ML algorithm to learn and produce right predictions. (70% of dataset is training dataset)

The test dataset, however, is used to assess how well ML algorithm is trained with the training dataset. You can’t simply reuse the training dataset in the testing stage because ML algorithm will already “know” the expected output, which defeats the purpose of testing the algorithm. (30% of dataset is testing dataset)

# Model Comparison & Selection


The evaluation metric is set based on **R^2 (coefficient of determination)** regression score function, that will represents the proportion of the variance for items (crops) in the regression model. **R^2** score shows how well terms (data points) fit a curve or line.

**R^2** is a statistical measure between 0 and 1 which calculates how similar a regression line is to the data it’s fitted to. If it’s a 1, the model 100% predicts the data variance; if it’s a 0, the model predicts none of the variance.

From results viewd above, **Decision Tree Regressor** has the highest R^2 score 0f **96%**, **GradientBoostingRegressor** comes second.


 I'll also calculate **Adjusted R^2** also indicates how well terms fit a curve or line, but adjusts for the number of terms in a model. If you add more and more useless variables to a model, adjusted r-squared will decrease. If you add more useful variables, adjusted r-squared will increase.
Adjusted R2 will always be less than or equal to R2.

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the Random Forest model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(train_data, train_labels)

# Generate predictions from the Random Forest model to use as features for the DNN
rf_predictions_train = rf_model.predict(train_data)
rf_predictions_test = rf_model.predict(test_data)
```

```python
import numpy as np

# Extend the original datasets with the Random Forest predictions
dnn_train_data = np.column_stack((train_data, rf_predictions_train))
dnn_test_data = np.column_stack((test_data, rf_predictions_test))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l1_l2

# Define the DNN architecture with Dropout and Regularization
dnn_model = Sequential([
    Dense(128, activation='relu', input_shape=(dnn_train_data.shape[1],), kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
    BatchNormalization(),
    Dense(128, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
    BatchNormalization(),
    Dropout(0.3),
    Dense(1)
])
```

```python
from tensorflow.keras.optimizers import Adam

# Compile the model
dnn_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model with the training data and early stopping
dnn_model.fit(dnn_train_data, train_labels, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

# Hyper Parameter Tuning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from tqdm import tqdm

def manual_hyperparameter_tuning(train_data, train_labels):
    # Define a set of hyperparameters to test
    n_estimators_options = [100, 200, 300]
    max_depth_options = [10, 20, None]
    min_samples_split_options = [2, 5, 10]
    min_samples_leaf_options = [1, 2, 4]

    best_score = float('inf')
    best_params = {}

    # Prepare combinations for tqdm
    total_combinations = [(n, md, mss, msl) for n in n_estimators_options
                                             for md in max_depth_options
                                             for mss in min_samples_split_options
                                             for msl in min_samples_leaf_options]

    # Iterate over combinations with tqdm progress bar
    for n_estimators, max_depth, min_samples_split, min_samples_leaf in tqdm(total_combinations, desc="RF Tuning Progress"):
        # Create and train the Random Forest model
        model = RandomForestRegressor(
            n_estimators=n_estimators,
            max_depth=max_depth,
            min_samples_split=min_samples_split,
            min_samples_leaf=min_samples_leaf,
            random_state=42
        )
        model.fit(train_data, train_labels)

        # Evaluate the model
        scores = cross_val_score(model, train_data, train_labels, cv=3, scoring='neg_mean_squared_error')
        score = -scores.mean()  # Convert from negative MSE to MSE

        # Update the best parameters if this model is better
        if score < best_score:
            best_score = score
            best_params = {
                'n_estimators': n_estimators,
                'max_depth': max_depth,
                'min_samples_split': min_samples_split,
                'min_samples_leaf': min_samples_leaf
            }

    return best_params
```

```python
from tensorflow.keras.callbacks import EarlyStopping

def tune_dnn(train_data, train_labels, epochs=10):
    best_mse = float('inf')
    best_model = None
    best_params = {}

    # Prepare the combinations
    hyperparameters = [(units, dropout_rate, learning_rate)
                       for units in [64, 128, 256]
                       for dropout_rate in [0.2, 0.3]
                       for learning_rate in [0.001, 0.01]]

    # Early stopping callback
    early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

    for units, dropout_rate, learning_rate in tqdm(hyperparameters, desc="DNN Tuning Progress"):
        model = Sequential([
            Dense(units, activation='relu', input_shape=(train_data.shape[1],), kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
            BatchNormalization(),
            Dropout(dropout_rate),
            Dense(units // 2, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
            BatchNormalization(),
            Dense(units, activation='relu', kernel_regularizer=l1_l2(l1=0.01, l2=0.01)),
            BatchNormalization(),
            Dropout(dropout_rate),
            Dense(1)
        ])

        model.compile(optimizer=Adam(learning_rate=learning_rate), loss='mean_squared_error')
        model.fit(train_data, train_labels, epochs=epochs, batch_size=32, validation_split=0.2, verbose=0, callbacks=[early_stopping])
        predictions = model.predict(train_data)
        mse = mean_squared_error(train_labels, predictions)

        if mse < best_mse:
            best_mse = mse
            best_model = model
            best_params = {'units': units, 'dropout_rate': dropout_rate, 'learning_rate': learning_rate}

    return best_params
```
